```python
import pandas as pd
import numpy as np

## Reading the csv file
bank_marketing = pd.read_csv('bank-additional-full.csv', sep = ';')
bank_marketing.head()
```

```python
## Creating the conversion variables
bank_marketing['conversion'] = np.where(bank_marketing['y'] == 'yes', 1, 0)
bank_marketing.head()
```

```python
## Computing the conversion rate (11.27%)
round(100*bank_marketing['conversion'].value_counts() / bank_marketing.shape[0], 2)
```

```python
## Computing conversion by education
round(100*bank_marketing.groupby('education')['conversion'].mean(), 2)
```

```python
## Computing conversion by job
round(100*bank_marketing.groupby('job')['conversion'].mean().sort_values(ascending = False), 2)
```
