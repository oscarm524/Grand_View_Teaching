```python
# Wins = 71.87 + 0.101 × Payroll − 0.060 × League

# 1a. Slope interpretation. For teams on the same league, for 1 million of increase in 
# payroll, on average, wins increases by 0.101.

# 1b. Intercept = 71.87 - 0.06 x 1 = 71.81

# 1c. National League team => League = 0 => Wins = 71.87 + 0.101 x 98 = 81.768
```

```python
import pandas as pd 
import numpy as np

import statsmodels.formula.api as smf

# Reading the  data 
nba = pd.read_csv('NBA_2006_2007.csv')
nba.head()
```

```python
# Building the linear model 
lm_md = smf.ols(formula = 'W ~ EFG_pct + TOV_pct + OREB_pct + FTA_rate', data = nba).fit()

# Extracting model results
lm_md.summary()
```

```python
new_obs = pd.DataFrame({'EFG_pct': [51],
                        'TOV_pct': [16],
                        'OREB_pct': [32], 
                        'FTA_rate': [0.35]})
new_obs
```

```python
# Predicting win
lm_md.predict(new_obs)
```
