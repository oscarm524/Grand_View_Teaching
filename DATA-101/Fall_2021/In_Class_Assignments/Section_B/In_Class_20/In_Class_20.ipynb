```python
import pandas as pd 

## Reading the csv file 
autos = pd.read_csv('autos.csv')
autos.head()
```

```python
## Defining the input and target variables 
X = autos[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]
Y = autos['mpg']

from sklearn.model_selection import train_test_split

## Splitting th data 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor 

## Defining the number of trees
n_tree = [100, 300, 500, 800, 1000]

## Extracting the trees 
n = len(n_tree)

## Defining the data-frame to store results
results = pd.DataFrame({'Number_of_Trees': np.repeat(np.nan, n), 'RMSE': np.repeat(np.nan, n), 'MAE': np.repeat(np.nan, n)})

for i in range(0, n):
    
    ## Building the random forest model 
    RF = RandomForestRegressor(n_estimators = n_tree[i]).fit(X_train, Y_train)
    
    ## Predicting on the test dataset 
    RF_pred = RF.predict(X_test)
    
    ## Storing results
    results['Number_of_Trees'][i] = n_tree[i]
    results['RMSE'][i] = np.sqrt(np.mean((RF_pred - Y_test)**2))
    results['MAE'][i] = np.mean(abs(RF_pred - Y_test))
    
results
```
