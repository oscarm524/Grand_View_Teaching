```python
import pandas as pd 
import matplotlib.pyplot as plt

autos = pd.read_csv('Automobile_data.csv')
autos.head()
```

```python
plt.plot('horsepower', 'price', 'o', data = autos)
plt.xlabel('Horsepower')
plt.ylabel('Car Price');
```

```python
import statsmodels.formula.api as smf

# Building the linear regression model 
lm_md = smf.ols(formula = 'price ~ horsepower', data = autos).fit()

# Extracting model results 
lm_md.summary()
```

```python
# Slope interpretation: For one unit of increase in horsepower, on average, the car 
# price increases by 208.67.
```
