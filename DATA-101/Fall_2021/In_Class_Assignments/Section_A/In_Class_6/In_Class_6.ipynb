```python
import pandas as pd

## Reading the csv file 
Index = pd.read_csv('Index.csv')
Index.head()
```

```python
## Creating a bar chart for Gender
Index['Gender'].value_counts().plot(kind = 'bar')
```

```python
## From the above bar chart, we see that the female bar is a little higher than the male bar
```

```python
## Creating the histogram of Height
Index['Height'].plot(kind = 'hist')
```

```python
## The distibution is pretty flat with two peaks: one around 165 and another around 185
```

```python
import matplotlib.pyplot as plt 

## Creating the scatter plot 
plt.plot(Index['Height'], Index['Weight'], 'o')
plt.grid()
plt.show()
```

```python
## From the above plot, we don't see a clear pattern between height and weight
```
