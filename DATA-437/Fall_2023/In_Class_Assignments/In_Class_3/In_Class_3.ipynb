```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt; plt.style.use('ggplot')

# Reading the csv file 
goalie_stats = pd.read_csv('game_goalie_stats.csv')
goalie_stats.head()
```

```python
# Computing the average time of ice 
goalie_stats.groupby('player_id')['timeOnIce'].mean()
```

```python
pd.DataFrame(goalie_stats.groupby('player_id')['timeOnIce'].sum()).sort_values(by = 'timeOnIce')
```

```python
pd.DataFrame(goalie_stats.groupby('player_id')['timeOnIce'].sum()).sort_values(by = 'timeOnIce', ascending = False)
```

```python
goalie_stats.groupby('player_id')['game_id'].nunique()
```

```python
goalie_stats.groupby('player_id')['game_id'].count()
```

```python
goalie_stats.groupby('player_id')['game_id'].count().hist()
plt.xlabel('Number of Games')
plt.ylabel('Frequency');
```

```python
# From the above, the distribution of the number of games is right-skewed.
```
