```python
import pandas as pd
import statsmodels.formula.api as smf

## Reading the csv file 
autos = pd.read_csv('Automobile_data.csv')
autos.head()
```

```python
## Building the linear regression model 
lm_md = smf.ols(formula = 'price ~ horsepower + length', data = autos).fit()

## Extracting the model results
lm_md.summary()
```

```python
## Predicting the price 
new_obs = pd.DataFrame({'horsepower': [150], 'length': [170]}) 

lm_md.predict(new_obs)
```
