```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt; plt.style.use('ggplot')

# Reading the csv file
baseball = pd.read_csv('standings.csv')
baseball.head()
```

```python
plt.plot('Runs Scored', 'Wins', 'o', data = baseball, color = 'steelblue')
plt.xlabel('Runs Scored')
plt.ylabel('Wins');
```

```python
# From the above chart, we see there is a positive linear relationship between 
# runs scored and wins.
```

```python
r = baseball[['Runs Scored', 'Wins']].corr()
r
```

```python
r = 0.655213
ME = 2*np.sqrt((1 - r**2) / baseball.shape[0])
print(f"The ME of the correlation is {ME}")
```

```python
# By comparing r to the ME, we conclude that the correlation between Runs Scored and 
# Wins is significant.
```
