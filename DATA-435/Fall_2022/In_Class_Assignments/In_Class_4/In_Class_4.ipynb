```python
import pandas as pd

## Reading the csv file 
bank_marketing = pd.read_csv('bank-additional-full.csv', sep = ';')
bank_marketing.head()
```

```python
import numpy as np

## Changing labels to numbers
bank_marketing['conversion'] = np.where(bank_marketing['y'] == 'no', 0, 1)
bank_marketing.head()
```

```python
import statsmodels.formula.api as smf

## Building the logistic regression model 
logit_md = smf.logit('conversion ~ age + duration', data = bank_marketing).fit()

## Extracting model results
logit_md.summary()
```

```python
## From the model output, we see that both age and duration have a positive relationship
## with conversion. The older the customer is, the more likely to convert. The longer the 
## call last, the more likely to convert.
```
