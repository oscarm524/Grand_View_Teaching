```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt; plt.style.use('ggplot')

# Reading the csv file
Index = pd.read_csv('Index.csv')
Index.head()
```

```python
Index['Gender'].value_counts().plot(kind = 'bar', color = 'steelblue')
plt.xlabel('Gender')
plt.ylabel('Frequency');
```

```python
Index['Height'].hist(color = 'steelblue')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency');
```

```python
plt.plot('Height', 'Weight', 'o', data = Index, color = 'steelblue')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)');
```
