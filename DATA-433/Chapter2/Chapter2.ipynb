```python
## Importing yahoo finance
import yfinance as yf 

## Importing Apple historical stock data 
apple = yf.download('AAPL', start = '2018-01-01', end = '2020-10-31')

## Computing the moving averages
apple['MA_50'] = apple['Close'].rolling(window = 50).mean()
apple['MA_200'] = apple['Close'].rolling(window = 200).mean()
```

```python
## Importing yahoo finance
import yfinance as yf

## Downloading Amazon stock price data 
amazon = yf.download('AMZN', start = '2019-01-01', end = '2020-12-31')

## Computing the moving averages and standard deviation
amazon['ma'] = amazon['Close'].rolling(window = 20).mean()
amazon['sd'] = amazon['Close'].rolling(window = 20).std(doof = 1)

## Computing the Bollinger bands
amazon['up'] = amazon['ma'] + 2*amazon['sd']
amazon['low'] = amazon['ma'] - 2*amazon['sd']
```

```python
## Importing matplotlib
import matplotlib.pyplot as plt

## Plotting the Bollinger bands
plt.plot('Close', data = amazon, label = 'Close Price', color = 'black')
plt.plot('ma', data = amazon, label = '20-day MA', color = 'blue')
plt.plot('up', data = amazon, label = 'Upper Band', color = 'green')
plt.plot('low', data = amazon, label = 'Lower Band', color = 'green')
plt.xlabel('Date')
plt.ylabel('Close Price ($)')
plt.legend(loc = 'upper left')
plt.xticks(rotation = 45)
plt.grid()
```

```python
## Importing yahoo finance
import yfinance as yf

## Downloading Amazon stock price data 
amazon = yf.download('AMZN', start = '2019-01-01', end = '2020-12-31')

## Importing pandas and numpy
import pandas as pd
import numpy as np

## Selecting variable of interest
amazon_close = pd.DataFrame(amazon['Close'])

## Calculating difference in price
amazon_close['Delta'] = amazon_close['Close'].diff()

## Computing dummy variables to indicate whether price went up or down
amazon_close['up'] = np.where(amazon_close['Delta'] > 0, 1, 0)
amazon_close['down'] = np.where(amazon_close['Delta'] <= 0, 1, 0)

## Dropping missing values
amazon_close = amazon_close.dropna(axis = 0)

## Computing prices for up and down 
amazon_close['up_val'] = abs(amazon_close['Delta'] * amazon_close['up'])
amazon_close['down_val'] = abs(amazon_close['Delta'] * amazon_close['down'])

## Computing up and down averages
amazon_close['up_avg'] = amazon_close['up_val'].rolling(window = 14).mean()
amazon_close['down_avg'] = amazon_close['down_val'].rolling(window = 14).mean()

## Computing the up and down Wilder Moving Averages
amazon_close['up_WMA'] = amazon_close['up_avg']
amazon_close['down_WMA'] = amazon_close['down_avg']

for i in range(14, amazon_close.shape[0]):

    amazon_close['up_WMA'][i] = amazon_close['up_WMA'][i-1] + (amazon_close['up_avg'][i] - amazon_close['up_WMA'][i-1]) / 14
    amazon_close['down_WMA'][i] = amazon_close['down_WMA'][i-1] + (amazon_close['down_avg'][i] - amazon_close['down_WMA'][i-1]) / 14

## Calculating the RS
amazon_close['RS'] = amazon_close['up_WMA'] / amazon_close['down_WMA']

## Calculating the RIS
amazon_close['RSI'] = 100 - (100/ (1 + amazon_close['RS']))
```

```python
## Visualizing the RSI
import matplotlib.pyplot as plt

plt.plot('RSI', data = amazon_close, color = 'black')
plt.axhline(y = 30, color = 'black', linestyle = '--')
plt.axhline(y = 70, color = 'black', linestyle = '--')
plt.ylabel('RSI (14-day Moving Average)')
plt.xlabel('Date')
plt.xticks(rotation = 45)
plt.grid()
```
