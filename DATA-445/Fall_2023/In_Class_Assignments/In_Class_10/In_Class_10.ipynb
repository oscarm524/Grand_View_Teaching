```python
import pandas as pd
import numpy as np

from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from sklearn.neural_network import MLPRegressor

# Reading the csv
car_price = pd.read_csv('CarPrice_Assignment.csv')
car_price.head()
```

```python
X = car_price[['wheelbase', 'enginesize', 'compressionratio', 'horsepower', 'peakrpm', 'citympg']].copy()
Y = car_price['price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
md1 = make_pipeline(MinMaxScaler(),
                    MLPRegressor(hidden_layer_sizes = (10,),
                                 activation = 'relu', 
                                 solver = 'adam', 
                                 max_iter = 100000,
                                 batch_size = 100)).fit(X_train, Y_train)

# Predicting on test
md1_pred = md1.predict(X_test)
md1_mse = mean_squared_error(Y_test, md1_pred)
print('The model 1 mse is', md1_mse)
```

```python
md2 = make_pipeline(MinMaxScaler(),
                    MLPRegressor(hidden_layer_sizes = (10, 8),
                                 activation = 'relu', 
                                 solver = 'adam', 
                                 max_iter = 100000,
                                 batch_size = 100)).fit(X_train, Y_train)

# Predicting on test
md2_pred = md2.predict(X_test)
md2_mse = mean_squared_error(Y_test, md2_pred)
print('The model 2 mse is', md2_mse)
```
