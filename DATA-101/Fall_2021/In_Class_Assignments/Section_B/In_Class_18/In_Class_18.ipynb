```python
import pandas as pd 

## Reading the  csv file
churn = pd.read_csv('Customer_Churn.csv')
churn.head()
```

```python
import numpy as np

## Changing labels to 0-1
churn['Churn_numb'] = np.where(churn['Churn'] == 'No', 0, 1)
churn.head()
```

```python
## Defining the input and target variables 
X = churn[['SeniorCitizen', 'tenure', 'MonthlyCharges']]
Y = churn['Churn_numb']

from sklearn.model_selection import train_test_split

## Splitting the data 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y)
```

```python
## Logistic regression 
from sklearn.linear_model import LogisticRegression

## Building the logistic model 
logit_md = LogisticRegression().fit(X_train, Y_train)

## Predicting in the test dataset 
logit_pred = logit_md.predict_proba(X_test)[:, 1]
logit_pred_label = np.where(logit_pred < 0.35, 0, 1)

from sklearn.metrics import confusion_matrix, accuracy_score, recall_score

print('The confusion matrix is', confusion_matrix(Y_test, logit_pred_label))
print('The accuracy is', accuracy_score(Y_test, logit_pred_label))
print('The recall is', recall_score(Y_test, logit_pred_label))
```

```python
## Random forest
from sklearn.ensemble import RandomForestClassifier

## Building the random forest model 
RF_md = RandomForestClassifier().fit(X_train, Y_train)

## Predicting in the test dataset 
RF_pred = RF_md.predict_proba(X_test)[:, 1]
RF_pred_label = np.where(RF_pred < 0.35, 0, 1)

print('The confusion matrix is', confusion_matrix(Y_test, RF_pred_label))
print('The accuracy is', accuracy_score(Y_test, RF_pred_label))
print('The recall is', recall_score(Y_test, RF_pred_label))
```
