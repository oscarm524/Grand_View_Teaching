```python
import pandas as pd

## Reading csv file 
batting = pd.read_csv('Batting.csv')
batting.head()
```

```python
## Aggregating data at the player level
batting_agg = batting.groupby('playerID')[['HR', 'AB', 'SO']].sum()
batting_agg.head()
```

```python
## Selecting players with at least 5000 at bats
batting_agg = batting_agg[batting_agg['AB'] >= 5000]
batting_agg.head()
```

```python
## Computing home-runs and strikeout rates
batting_agg['HR_rate'] = batting_agg['HR'] / batting_agg['AB']
batting_agg['SO_rate'] = batting_agg['SO'] / batting_agg['AB']
batting_agg.head()
```

```python
import matplotlib.pyplot as plt 

## Creating the scatter-plot between home-run rate and strikeout rate
plt.plot('HR_rate', 'SO_rate', 'o', data = batting_agg)
plt.xlabel('Home-Run Rate')
plt.ylabel('Strikeout Rate')
plt.grid()
plt.show()

## From the scatter-plot, it seems that there is a positive linear 
## relationship.
```

```python
import statsmodels.formula.api as smf

## Building the linear regression model
lm_md = smf.ols(formula = 'SO_rate ~ HR_rate', data = batting_agg).fit()

## Extracting model results
lm_md.summary()
```

```python
## Defining new data-points
newdata = pd.DataFrame({'HR_rate': [0.05]})

## Predicting HR_rate on the new data-points
lm_md.predict(newdata)
```
