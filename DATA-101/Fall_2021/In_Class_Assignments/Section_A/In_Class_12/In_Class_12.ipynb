```python
import pandas as pd

## Reading the csv file 
autos = pd.read_csv('Automobile_data.csv')
autos.head()
```

```python
import statsmodels.formula.api as smf

## Building the linear regression model 
lm_md = smf.ols(formula = 'price ~ horsepower + length', data = autos).fit()

## Extracting model results
lm_md.summary()
```

```python
## Defining the new dataset
newdata = pd.DataFrame({'horsepower': [150], 'length': [170]})

## Predicting price for the new dataset
lm_md.predict(newdata)
```
